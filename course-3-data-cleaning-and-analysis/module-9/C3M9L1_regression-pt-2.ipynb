NORMAL QUANTILE PLOTS

This is an example of how to draw a picture of a binomial distribution using the dbinom function

```r
# Parameters
n <- 10
p <- 0.83
x_vals <- 0:n

# Compute probabilities
prob_vals <- dbinom(x_vals, size = n, prob = p)

# Plot the PMF
barplot(prob_vals,
        names.arg = x_vals,
        col = "royalblue",
        main = paste("Binomial(n =", n, ", p =", p, ")"),
        xlab = "Number of Successes",
        ylab = "Probability",
        ylim = c(0, max(prob_vals) * 1.1))
```

Exponential Distribution picture



```r

# Parameters
lambda <- 1 / 60
x_vals <- seq(0, 500, length.out = 1000)

# Plot the density function
plot(x_vals, dexp(x_vals, rate = lambda), type = "l", col = "blue", lwd = 3,
     main = "Exponential Density Function (mean = 60)", xlab = "x", ylab = "Density")
abline(h = 0, col = 'lightgrey')
abline(v = 0, col = 'lightgrey')
```

Normal Distribution picture

```r
x_vals <- seq(-3, 3, length.out = 1000)

# Plot the density function
plot(x_vals, dnorm(x_vals), type = "l", col = "blue", lwd = 3,
     main = "Normal Density Function (mean = 0, sd = 1)", xlab = "x", ylab = "Density")
abline(h = 0, col = 'lightgrey')
```

Dissolved Oxygen in Bridgeport Harbor

```r
#Bridgeport Harbor Dissolved Oxygen Data
dissox <- read.csv("https://raw.githubusercontent.com/jreuning/YSE_EDA_data/refs/heads/main/BridgeportDO.csv", header = T)
names(dissox)
```

```r
hist(dissox$DissolvedOxygen, col = 'blue', main = 'Dissolved Oxygen Levels', xlab = 'Dissolved oxygen, mg/l', breaks = 15)
```

```r
library(car)

#Normal Quantile Plot
qqp(dissox$DissolvedOxygen, col = 'blue', main = "Dissolved Oxygen Levels",  
    ylab = "Dissolved Oxygen, mg/l", 
    xlab = "Normal Quantiles",
    pch = 19)
```

Simulation to show the effect of sample size on linearity in a normal quantile plot

```r
par(mfrow = c(2,2), cex.main = .8)

vec <- c(5, 10, 50, 500)
for (i in vec){
  qqp(rnorm(i), main = paste("Random Normal Data, N=", i), pch = 19, col = 'blue')
}
```
