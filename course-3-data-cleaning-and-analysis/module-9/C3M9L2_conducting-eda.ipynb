SCATTERPLOTS and CORRELATION We start by looking at size measures of cherry trees

```r
#Cherry Tree Data
cherry <-  read.csv("https://raw.githubusercontent.com/jreuning/YSE_EDA_data/refs/heads/main/Cherry.Trees.csv", header = T)
head(cherry)
```

```r
#Make a scatterplot
plot(cherry$Diameter, cherry$Volume, 
     pch = 19, 
     xlab = "Diameter", 
     ylab = "Volume", 
     main = "Scatterplot of Volume vs Diameter", 
     col = 'blue')
```

```r
# Here is equivalent syntax
plot(cherry$Volume ~ cherry$Diameter, 
     pch = 19, 
     xlab = "Diameter", 
     ylab = "Volume", 
     main = "Scatterplot of Volume vs Diameter", 
     col = 'red')
```

Let's calculate some correlations

```r
#we use the cor() function
cor(cherry$Volume, cherry$Diameter)
#fewer decimal places
round(cor(cherry$Volume, cherry$Diameter), 2)
#notice order doesn't matter
round(cor(cherry$Diameter, cherry$Volume), 2)
```

If we have more than two variables, we get all pairwise correlations

```r
round(cor(cherry), 2)
#note that by definition, the correlation of a variable with itself is always 1
```

I was curious that height and diameter were less highly correlated.  Look at a scatterplot

```r
plot(cherry$Height ~ cherry$Diameter, 
     pch = 19, 
     xlab = "Diameter", 
     ylab = "Volume", 
     main = "Scatterplot of Volume vs Diameter", 
     col = 'red')
```

Next we use the World Bank Data from 2016 to look at several relationships

```r
#Positive Association:
WB <- read.csv("https://raw.githubusercontent.com/jreuning/sds230_data/refs/heads/main/WB.2016.csv", header = T, as.is = T)

plot(WB$Rural, WB$Fertility16, 
     pch = 19, cex.main = 1, 
     xlab = "Rural Pop Pct", 
     ylab = "Fertility Rate 2016", 
     main = "Scatterplot of Fertility Rate 2016 vs Rural Pop Pct", 
     col = 'blue')

plot(WB$EnergyUse, WB$CO2, 
     pch = 19, 
     xlab = "Energy Use per capita", 
     ylab = "CO2 Emissions per capita", 
     main = "Scatterplot of Energy Use per capita vs CO2 Emissions per capita", 
     col = 'blue')
```

```r
#No Association:
plot(WB$FossilPct, as.numeric(WB$Deforestation),
     pch = 19, 
     xlab = "Pct Fossil Fuels", 
     ylab = "Deforestation Rate", 
     main = "Scatterplot of Pct Fossil Fuels vs. Deforestation Rate", 
     col = 'blue')
```

```r
#Negative Association:
plot(WB$GNI, WB$PM2.5, 
     pch = 19, 
     cex.main = 1, 
     xlab = "GNI per capita", 
     ylab = "PM 2.5 rate", 
     main = "Scatterplot of GNI vs PM 2.5", 
     col = 'blue')
```

Let's look at body brain data in mammals

```r
#get data
BodyBrain <- read.csv("https://raw.githubusercontent.com/jreuning/YSE_EDA_data/refs/heads/main/BodyBrainMammals.csv", header = T)
head(BodyBrain)
```

```r
#Make a scatterplot
plot(BodyBrain$BodyWt, BodyBrain$BrainWt, 
     pch = 19,
     cex = 1.3,
     xlab = "Body Weight (kg)", 
     ylab = "Brain Weight(g)", 
     cex.main = 1, 
     main = "Scatterplot of Body Weight(kg) vs Brain Weight(g)", 
     col = 'red')
mtext(paste("Correlation =", round(cor(BodyBrain$BodyWt, BodyBrain$BrainWt), 2)))
```

```r
Examine this now on the natural log scale
```

```r
#Now let's plot the log of both variables:

plot(BodyBrain$LnBodyWt, BodyBrain$LnBrainWt, 
     pch = 19, 
     xlab = "LN(Body Weight)", 
     ylab = "LN(Brain Weight)", 
     cex.main = 1, 
     main = "Scatterplot of LN(Brain Weight) vs LN(Body Weight)", 
     col = 'red')
mtext(paste("Correlation =", round(cor(BodyBrain$LnBodyWt, BodyBrain$LnBrainWt), 2)))
```

Some Examples of the dangers of calculating correlation without making plots!

```r
#Quick function to make a scatterplot and report correlation
myCor <- function(x, y){
  #make the scatterplot
  plot(x, y, pch = 19, col = "red")
  #add text in the margin reporting the sample correlation)
  mtext(paste("Sample Correlation =", round(cor(x, y), 3)), cex = 1.2)
}
```

```r
# CORRELATION NOT APPROPRIATE!
# Fit a parabola
x <- c(-30:70) * .05
y <- -x^2 + rnorm(101)
myCor(x,y)
```

```r
# CORRELATION NOT APPROPRIATE!
# Fit a  . . . Smack
x <- seq(0, pi, .01)
y <- sin(x)
x <- c(x, x) 
y <- c(y, -y) + rnorm(length(x))/10
myCor(x, y)
```

```r
# CORRELATION NOT APPROPRIATE!
# Use ANCOVA (i.e. fit two lines) - we'll do this later in the course
x <- c(1:100)*.1
y <- -x + rnorm(100)
x <- c(x, x + 10)
y <- c(y, y + 10)
myCor(x, y)
```

Finally, an example of the effect of an outlier on correlation

```r
#Outlier make correlation artificially high
x <- c(rnorm(100), 20)
y <- c(rnorm(100), 20)
myCor(x, y)

#without the point (20,20)
myCor(x[-101], y[-101])
```

```r
#Outlier make correlation artificially low AND goes the wrong direction!
x <- c(1:10, 20)
y <- c(10:1, 20)
myCor(x, y)
```
